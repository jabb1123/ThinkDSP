Question 2

Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly.

```python
class SawtoothChirp(thinkdsp.Chirp):
    def _evaluate (self,ts,freqs):
        dts = numpy.diff(ts)
        dps = numpy.pi*2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        frac, _ = numpy.modf(phases/(numpy.pi*2))
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys
```

```python
signal = SawtoothChirp()
wave1 = signal.make_wave(duration=2)
spectrogram = wave1.make_spectrogram(512)
spectrogram.plot()
wave1.make_audio()
```

Question 3

Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
def Glissando(bottom = 440, top = 880, amp =1):
    signal1 = thinkdsp.Chirp(bottom,top,amp).make_wave(duration=1)
    signal2 = thinkdsp.Chirp(top,bottom,amp).make_wave(duration=1)
    mix = signal1|signal2
    return mix
wave = Glissando(262,349,1)

wave.make_audio()
```

Make or find a recording of a series of vowel sounds and look at the spectrogram. Can you identify different vowels?

```python
wave = thinkdsp.read_wave('Vowels.wav')
wave.framerate*=2
wave.plot()
wave.make_audio()
```

```python
spectrogram = wave.make_spectrogram(seg_length=1024)
spectrogram.plot(high=70)
thinkplot.config(xlabel='time(s)', ylabel='frequency (Hz)', legend=False)
```
